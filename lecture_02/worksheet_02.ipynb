### Effective Programming

a) What is a drawback of the top down approach?

The top-down approach involves breaking down a problem into smaller and manageable sub-problems and then solving them one by one. However, there are some drawbacks by taking this approach:

a) inflexibility
b) Limited detail at the top
c) Difficulty in understanding lower-level details
d) Increased time for initial planing 

b) What is a drawback of the bottom up approach?

The bottom-up approach involves building a system from the smallest components without having a clear understanding of the oveerall system architecture. However, There are a few drawbacks when taking this approach:

a) Lack of Holistic view
b) integration challenges
c) Difficulty in managing complexity
d) Potential redundancy

a) Complete the code below to read in a filepath to the `train.csv` and returns the DataFrame.

```python
import pandas as pd

df = pd.read_csv('train.csv')
df.describe()
```

b) Complete the code so it returns the number of rows that have at least one empty column value

```python

num_row = df[df.isnull().any(axis=1)].shape[0]
#print(num_row)
print("there are " +  str(num_row) + " rows with at least one empty value")


#df.info()
df.dtypes
```

c) Complete the code below to remove all columns with more than 200 NaN values

```python
threshold = 200
df = df.dropna(axis=1, thresh=len(df) - threshold)
#print(df.column)
df.describe()
```

d) Complete the code below to replaces `male` with 0 and `female` with 1

```python
df['Sex'] = df['Sex'].replace({'male': 0, 'female': 1})

df.head()
```

e) Complete the code below to add four columns `First Name`, `Middle Name`, `Last Name`, and `Title` corresponding to the value in the `name` column.

For example: `Braund, Mr. Owen Harris` would be: 

|First Name | Middle Name | Last Name | Title |
|-----------|-------------|-----------|-------|
| Owen      |  Harris     |  Braund   | Mr    |

Anything not clearly one of the above 4 categories can be ignored.

```python
df[['First Name', 'Middle Name', 'Last Name', 'Title']] = df['Name'].str.extract(r'(?P<Last_Name>[^,]+), \s(?P<Title>[^.]+)\.\s(?P<First_Name>[^\s]+)\s(?P<Middle_Naame>.+)')
df.head()
```

f) Complete the code below to replace all missing ages with the average age

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.head()
```

g) Plot a bar chart of the average age of those that survived and did not survive. Briefly comment on what you observe.

```python
import matplotlib.pyplot as plt

df['survived_age'] = df[df['Survived'] == 1]['Age'].mean()
df['not_survived_age'] = df[df['Survived'] == 0]['Age'].mean()

plt.bar(['Survived', 'Not Survived'], ['survived_age', 'not_survived_age'])
plt.xlabel('Survival Status')
plt.ylabel('Average Age')
plt.title('Average Age of Survived and Not Survived Passengers')
plt.show()

df.head()
```
